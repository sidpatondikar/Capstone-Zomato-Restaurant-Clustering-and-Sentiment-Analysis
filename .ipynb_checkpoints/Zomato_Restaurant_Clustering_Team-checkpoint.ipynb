### Variables Description



**review_df**:

1. **Restaurant**: This column contains the name of the restaurant where the review was posted. It serves as a reference to the specific dining establishment under consideration.

2. **Reviewer**: In this column, you can find the name of the individual who posted the review. It helps identify the author of the feedback.

3. **Review**: This column provides the detailed review text given by the reviewer, offering insights into their dining experience, including comments and feedback about the restaurant.

4. **Rating**: The "Rating" column records the numeric evaluation provided by the reviewer, typically on a scale of 1 to 5. It quantifies the reviewer's overall satisfaction with the restaurant.

5. **Metadata**: This column includes additional information about the reviewer, such as the total number of reviews they have given on Zomato and the number of followers they have. This metadata can help assess the credibility and influence of the reviewer.

6. **Time**: The "Time" column indicates the date and time when the review was submitted. It offers a temporal reference point for analyzing trends and patterns in reviews.

7. **Pictures**: This column counts the number of images or pictures uploaded by the reviewer as part of their review. It provides insights into the visual content associated with the dining experience.

**restaurant_df**:

1. **Name**: The "Name" column displays the name of the restaurant. It serves as a unique identifier for each dining establishment listed in the dataset.

2. **Link**: In this column, you can find the hyperlink or URL that directs users to the restaurant's page on Zomato's platform. It facilitates easy access to additional information about the restaurant.

3. **Cost**: The "Cost" column indicates the average cost of a meal at the restaurant. It helps users gauge the affordability of dining at a particular establishment.

4. **Collections**: This column includes tags or categories that Zomato associates with the restaurant. These tags provide insights into the restaurant's style, cuisine, or unique features, helping users make informed dining choices.

5. **Cuisine**: The "Cuisine" column lists all the different types of cuisines available at the restaurant. It offers a comprehensive overview of the culinary offerings at the establishment.

6. **Timings**: In this column, you can find information about the operating hours of the restaurant. It specifies when the restaurant is open, allowing users to plan their visits accordingly.

````python
# Write your code to make your dataset analysis ready.

#Defining a function to get cuisines
def count_cuisines(dataframe):
    # Split the comma-separated cuisines and create a list of all cuisines
    all_cuisines = dataframe['cuisine'].str.split(', ').explode()

    # Use value_counts to count the occurrences of each cuisine
    cuisine_counts = all_cuisines.value_counts().reset_index()
    cuisine_counts.columns = ['Cuisine', 'Count']

    return cuisine_counts
````

````python
#Making a new dataframe for cuisines values
cuisine_counts = count_cuisines(restaurant_df)
cuisine_counts
````

````python
review_df.head()
````

````python
#Checking the ratings
review_df['Rating'].value_counts()
````

````python
review_df[review_df['Rating']=='Like']
````

Changing the Rating 'Like' to 5 based on the review

````python
#Changing all ratings to float values
def change_rating(X):
  if X=='Like':
    return float(5)
  else:
    return math.ceil(float(X))

review_df['Rating'] = review_df['Rating'].apply(change_rating)
````

````python
review_df['Rating'].value_counts()
````

````python
#Converting 'Time' column into datetime
review_df['Time'] = pd.to_datetime(review_df['Time'])
review_df['Time_HM'] =  review_df['Time'].dt.time

#Creating a new column to get hour of review
review_df["Hour"] = review_df['Time'].dt.hour
````

````python
#Creating a day column to get day of review
review_df['Day'] = review_df['Time'].dt.strftime('%A')
````

````python
review_df.head()
````

````python
restaurant_df.head()
````

````python
#Converting 'Cost' column into int
restaurant_df['cost'] = restaurant_df['cost'].apply(lambda x : int(x.replace(',','')))
restaurant_df.head()
````

````python
#Merging both dataframes to get restaurant ratings and costs
merged_restaurant = review_df[['Restaurant','Rating']]
merged_restaurant.rename(columns = {'Restaurant':'name'}, inplace=True)
merged_restaurant = merged_restaurant.merge(restaurant_df[['name','cost']], how='left',on='name')
````

````python
#Grouping on restaurant name to find average rating and cost of the restaurant
merged_restaurant = pd.DataFrame(merged_restaurant.groupby('name').mean()).reset_index()
merged_restaurant.sort_values('Rating',ascending=False, inplace=True)
merged_restaurant
````

````python
review_df.head()
````

````python
#Defining a function to get number of reviews made by a reviewer
def get_total_review(metadata):
  # Split the metadata string into parts
  metadata = metadata.split(',')
  review = metadata[0].split(' ')[0]
  return int(review)

#Defining a function for getting number of followers a reviewer has on Zomato
def get_follower(metadata):
  # Split the metadata string into parts
    parts = metadata.split(',')

    # Check if there are at least two parts (review and followers)
    if len(parts) >= 2:
        # Extract the second part (followers) and split it by whitespace
        followers_part = parts[1].strip()
        followers_parts = followers_part.split()

        # Check if there are at least two parts in the followers section
        if len(followers_parts) >= 2:
            # Attempt to extract the follower count
            follower = followers_parts[0]

            # Check if the extracted follower count is a valid number
            if follower.isdigit():
                return int(follower)

    # If any of the checks fail, return None (or any default value you prefer)
    return 0
````

````python
#Creating a new column for number of reviews
review_df['No_of_Reviews'] = review_df['Metadata'].apply(get_total_review)

#Creating a new columns for number of followers
review_df['Followers'] = review_df['Metadata'].apply(get_follower)
````

````python
review_df
````

````python
#Creating a new dataframe to get reviewer data
merged_reviewer = pd.DataFrame(review_df.groupby('Reviewer').agg({'No_of_Reviews': 'max', 'Followers': 'max', 'Rating': 'mean'}).reset_index())
merged_reviewer
````

##### What is/are the insight(s) found from the chart?

**Word Cloud Analysis of Restaurant Reviews**:

1. **Common Keywords - Good and Food**: Interestingly, the word clouds for both the top 10 highest rated restaurants and the bottom 10 rated restaurants prominently feature keywords like "good" and "food." This suggests that the taste of the food itself may not be the primary driver of ratings differentiation.

2. **Distinguishing Factors**: However, what sets apart highly-rated and lowly-rated restaurants are additional keywords like "ambience," "service," "quality," "staff," and "time of order." These words reflect the importance of factors beyond taste, such as the overall dining experience, service quality, staff behavior, and timeliness, in shaping customer perceptions and influencing their ratings.

These insights underline the multifaceted nature of restaurant reviews, where various aspects of the dining experience play a pivotal role in determining customer satisfaction and, consequently, restaurant ratings. Understanding these key differentiators can help restaurants and Zomato focus on areas that contribute to improved customer experiences and higher ratings.
